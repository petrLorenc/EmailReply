```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
result_without_pos = pickle.load(open("./result_without_pos.plk", mode="rb"))
result_pos = pickle.load(open("./result_pos.plk", mode="rb"))
result_glove = pickle.load(open("./result_glove.plk", mode="rb"))
result_glove_min_max = pickle.load(open("./results_glove_min_max.plk", mode="rb"))
result_doc2vec = pickle.load(open("./results_doc2vec.plk", mode="rb"))
```

```python
def showHistogram(results):
    tf_idf_scores = tuple(set([item[2] for item in results if item[1] == 'TfidfVectorizer']))
    BOW_scores = tuple(set([item[2] for item in results if item[1] == 'CountVectorizer']))

    names = tuple(set([item[0] for item in results]))
    # data to plot
    n_groups = max(len(tf_idf_scores), len(BOW_scores))

    # create plot
    index = np.arange(n_groups)
    bar_width = 0.35
    opacity = 0.8

    fig = plt.figure(figsize=(6, 8), dpi=100)


    rects1 = plt.bar(index, tf_idf_scores, bar_width,
                     alpha=opacity,
                     color='b',
                     label='Tf-idf')

    rects2 = plt.bar(index + bar_width, BOW_scores, bar_width,
                     alpha=opacity,
                     color='g',
                     label='Bag-of-words')

    plt.ylabel('Score')
    plt.title('Scores by vectorizer and model')
    plt.xticks(index + bar_width, names, rotation=70)

    plt.legend(loc="lower left")
    plt.tight_layout()
    plt.show()
    
def showHistogram2(results):
    scores = tuple([item[1] for item in results])

    names = tuple(set([item[0] for item in results]))
    
    fig = plt.figure(figsize=(6, 8), dpi=100)
    
    objects = names
    y_pos = np.arange(len(objects))
    performance = scores

    plt.bar(y_pos, performance, align='center', alpha=0.5)
    plt.xticks(y_pos, objects, rotation=70)
    
    plt.ylabel('Score')
    plt.title('Scores by vectorizer and model')

    plt.show()
```

```python
showHistogram(result_pos)
```

```python
showHistogram(result_without_pos)
```

```python
showHistogram(result_glove)
```

```python
showHistogram2(result_glove_min_max)
```

```python
showHistogram2(result_doc2vec)
```
